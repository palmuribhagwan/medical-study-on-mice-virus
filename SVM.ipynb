## Hyper Parameters

### c 
- what is c?

**regularization parameter. the strength of the regulrization is inversely proportional to C**

```python
model = SVC(kernel='linear' , C=10)
model.fit(x,y)
plot_svm(model,x,y)
```

```python
model = SVC(kernel='linear' , C=0.05)
model.fit(x,y)
plot_svm(model,x,y)
```

```python
model = SVC(kernel='rbf' , C=1)
model.fit(x,y)
plot_svm(model,x,y)
```

```python
model = SVC(kernel='sigmoid' )
model.fit(x,y)
plot_svm(model,x,y)
```
